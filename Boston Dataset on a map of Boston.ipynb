```python
import folium
from folium import plugins
import branca
import branca.colormap as cm
from IPython.display import display

#Initialise map and datapoints
Boston = folium.Map(location = [42.2, -70.955], zoom_start= 9.5, control_scale=True)
df = pd.read_csv("boston_corrected_corrected.csv")

#Initialise points-group and colourmap
Points = folium.FeatureGroup(name = 'Points')


colormap = cm.LinearColormap(colors=['black', 'darkred', 'red', 'yellow', 'green', 'darkgreen'],
                             index=[0, 10, 20, 30, 40, 50], vmin=0, vmax=50,
                             caption='Median value of owner-occupied homes (in $1000)')


#Add points in datapoints to points-group
for i in range(0, len(df)):
    Points.add_child(
        folium.Circle(
            location = [df.iloc[i]["LAT"], df.iloc[i]["LON"]], 
            color = colormap(df.iloc[i]["CMEDV"]),
            fill = True,
            lineColor = 'black',
            radius = 50,
            fill_opacity=0.7,
            popup = str("CMEDV: {}".format(df.iloc[i]["CMEDV"])))
    )

#Add points to map
Boston.add_child(Points)

#Add colourbar on map
Boston.add_child(colormap)

#Heatmap
location_data = df[['LAT', 'LON']]
location_data_low = location_data[df['CMEDV'] < 15]
location_data_high = location_data[df['CMEDV'] > 40]

location_data_high = [[row['LAT'],row['LON']] for index, row in location_data_high.iterrows()]
location_data_low = [[row['LAT'],row['LON']] for index, row in location_data_low.iterrows()]
#location_data_all = [[row['LAT'],row['LON']] for index, row in location_data.iterrows()]

# Plot heatmaps on the map
highgrad = {.33: 'lightgreen', .66: 'green', 1: 'darkgreen'}
lowgrad = {.33: 'yellow', .66: 'orange', 1: 'red'}
plugins.HeatMap(location_data_high, 
                gradient = highgrad, 
                min_opacity = 0.4, 
                name = 'Heatmap of high wealth areas CMEDV>$40000').add_to(Boston)
plugins.HeatMap(location_data_low, 
                gradient = lowgrad, 
                min_opacity = 0.4, 
                name = 'Heatmap of lower wealth areas CMEDV<$20000').add_to(Boston)
Boston.add_child(folium.map.LayerControl())

print("Please use the Layer Control button (below colourbar of points) \n to toggle heatmaps on")
display(Boston)
```
